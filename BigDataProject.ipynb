```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Dataset
The original dataset consists of 16,860 observations of motor vehicle collisions from 2006 to 2020 and is composed of 54 numerical and categorical features.

```python
file = 'Motor Vehicle Collisions with KSI Data.csv'
df=  pd.read_csv(file, sep=',');
```

```python
df
```

## Data Processing
Data is prepared to be applied to the models.

### Feature Selection
From the original 54 features, irrelevant and redundant variables were dropped, and new features were constructed.<br>
The following # features were chosen to be used in this study:
- feature 1
- feature 2
- feature 3
- feature 4
- ...
- feature n

```python
columns_to_drop = ['_id', 'ACCNUM', 'TIME', 'OFFSET', 'WARDNUM', 'DIVISION', 'ACCLOC', 'FATAL_NO', 'PEDTYPE', 'PEDACT', 'PEDCOND', 'CYCLISTYPE', 'CYCACT', 'CYCCOND', 'POLICE_DIVISION', 'ObjectId', 'geometry']
df = df.drop(columns=columns_to_drop)
df
```

## Class Distribution
The feature *'INJURY'* is the dependent feature that will classify the accident severity.<br>
 There are five types of injuries: None, Major, Minor, Minimal and Fatal.

```python
class_counts = df['INJURY'].value_counts()
classes = class_counts.index.tolist()
class_instance_count = class_counts.values
plt.bar(classes,class_instance_count, color='orange')
plt.title('Injury Severity Distribution')
plt.xlabel('Injury Type')
plt.ylabel('Number of Instances');
```
