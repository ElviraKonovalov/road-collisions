```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = 'Motor Vehicle Collisions with KSI Data.csv'
df=  pd.read_csv(file, sep=',');
```

```python
columns_to_drop = ['_id', 'ACCNUM', 'TIME', 'OFFSET', 'WARDNUM', 'DIVISION', 'ACCLOC', 'FATAL_NO', 'PEDTYPE', 'PEDACT', 'PEDCOND', 'CYCLISTYPE', 'CYCACT', 'CYCCOND', 'POLICE_DIVISION', 'ObjectId', 'geometry']
df = df.drop(columns=columns_to_drop)
```

### ROAD_CLASS

```python
print(df.ROAD_CLASS.unique()) # inital road_class categories
```

```python
df.ROAD_CLASS.isnull().sum() # how many nan values in road_class column
```

```python
'''
"Every street owned by the City of Toronto has been given one of five classifications (with the exception of public laneways)."
classes: Local road, Collector road, Minor arterial road, Major arterical road and Expressway
https://www.toronto.ca/services-payments/streets-parking-transportation/traffic-management/road-classification-system/about-the-road-classification-system/
'''
# replace nan values with most frequent value in the column
df['ROAD_CLASS'] = df['ROAD_CLASS'].fillna(df['ROAD_CLASS'].value_counts().index[0])
print(df.ROAD_CLASS.unique()) # road_class values without nan 
df.ROAD_CLASS.isnull().sum() # how many nan values after fillna
```

```python
# drop rows where ROAD_CLASS is pending
df.drop(df.index[df['ROAD_CLASS'] == 'Pending'], inplace=True)
print(df.ROAD_CLASS.unique())
```

```python
df.ROAD_CLASS.value_counts()
```

```python
# replacing Expressway Ramp (4 rows) with Expressway
df['ROAD_CLASS'] = df['ROAD_CLASS'].replace(['Expressway Ramp'],'Expressway')
# Major Arterial Ramp (1 row) with Major Arterial
df['ROAD_CLASS'] = df['ROAD_CLASS'].replace(['Major Arterial Ramp'],'Major Arterial')
df.ROAD_CLASS.value_counts()
```

```python
# get one hot encoding of columns ROAD_CLASS
hot = pd.get_dummies(df['ROAD_CLASS'])
# drop column ROAD_CLASS as it is now encoded
df = df.drop('ROAD_CLASS',axis = 1)
# join the encoded df
df = df.join(hot)
df
```

### VEHTYPE
Type of Vehicle

```python
df.VEHTYPE.value_counts()
```

```python
# replacing any type of Bus with Bus
df['VEHTYPE'] = df['VEHTYPE'].replace(['Bus (Other) (Go Bus, Gray Coach)', 'School Bus', 'Municipal Transit Bus (TTC)', 'Intercity Bus'], 'Bus')
# replacing any type of truck with Truck
df['VEHTYPE'] = df['VEHTYPE'].replace(['Truck-Tractor', 'Truck - Tank', 'Truck - Open', 'Truck - Dump', 'Truck - Closed (Blazer, etc)', 'Truck - Car Carrier', 'Truck (other)', 'Tow Truck', 'Pick Up Truck', ], 'Truck')
# replacing any type of emergency vehicle with Emergency
df['VEHTYPE'] = df['VEHTYPE'].replace(['Police Vehicle', 'Other Emergency Vehicle', 'Fire Vehicle'], 'Emergency')
# replacing any type of van with Van
df['VEHTYPE'] = df['VEHTYPE'].replace(['Delivery Van', 'Passenger Van'], 'Van')

df.VEHTYPE.value_counts()
```
